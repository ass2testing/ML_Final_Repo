### Introduction ###
In this project, my aim to build a machine learning model to predict median housing prices in California. 
The model could aid real estate agents, policymakers, and economic researchers in understanding housing market trends


### Problem Statement###

The challenge is to predict the median house value for California districts, based on various features that can influence house 
prices, such as median income, housing age, and location.


### Data Overview ###
i will use the California Housing dataset available from the `sklearn` dataset library. 
It contains demographic and geographical information along with median house values for districts across California.


### Import Libraries and Load the Data ###

Import necessary Python libraries and load the dataset.

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing

# Load the dataset
housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns=housing.feature_names)
df['MedHouseVal'] = housing.target
```
